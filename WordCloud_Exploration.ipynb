```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd
import os

os.chdir(r"E:\Download\Applied ML Project")
df = pd.read_csv("listings.csv")

df.head()
```

```python
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)
df['amenities'] = df['amenities'].str.replace('[^\w\s]', ' ')
df['description'] = df['amenities'].str.replace('[^\w\s]', ' ')
```

```python
text_amenities = df.loc[df['price'] > 500, 'amenities']

wordcloud = WordCloud().generate(''.join(text_amenities))
plt.imshow(wordcloud.recolor(random_state=1))
plt.title('Most mentioned amenities with price > 500/day')
plt.axis("off")
plt.show()

text_description = df.loc[df['price'] > 500, 'description']

wordcloud = WordCloud().generate(''.join(text_description))
plt.imshow(wordcloud.recolor(random_state=1))
plt.title('Most mentioned description with price > 500/day')
plt.axis("off")
plt.show()
```
